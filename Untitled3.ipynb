```python
import pandas as pd
import glob
import os

# =========================
# ⚙️ CONFIG
# =========================
DATA_PATH = "ECA_blended"
os.chdir(DATA_PATH)

# =========================
# 🔍 FIND DATA START
# =========================
def find_data_start(file):
    with open(file, 'r') as f:
        for i, line in enumerate(f):
            if line.strip().startswith("SOUID"):
                return i
    return None

# =========================
# 🌧️ LOAD RR
# =========================
def load_rr(file):
    start = find_data_start(file)
    df = pd.read_csv(file, skiprows=start)
    df.columns = df.columns.str.strip()

    df["DATE"] = pd.to_datetime(df["DATE"], format="%Y%m%d")
    df["year"] = df["DATE"].dt.year

    df["RR"] = pd.to_numeric(df["RR"], errors="coerce") / 10.0
    df = df[df["RR"] != -999.9]

    return df

# =========================
# 🌡️ LOAD TG
# =========================
def load_tg(file):
    start = find_data_start(file)
    df = pd.read_csv(file, skiprows=start)
    df.columns = df.columns.str.strip()

    df["DATE"] = pd.to_datetime(df["DATE"], format="%Y%m%d")
    df["year"] = df["DATE"].dt.year

    df["TG"] = pd.to_numeric(df["TG"], errors="coerce") / 10.0
    df = df[df["TG"] != -999.9]

    return df

# =========================
# 🌡️ LOAD TN
# =========================
def load_tn(file):
    start = find_data_start(file)
    df = pd.read_csv(file, skiprows=start)
    df.columns = df.columns.str.strip()

    df["DATE"] = pd.to_datetime(df["DATE"], format="%Y%m%d")
    df["year"] = df["DATE"].dt.year

    df["TN"] = pd.to_numeric(df["TN"], errors="coerce") / 10.0
    df = df[df["TN"] != -999.9]

    return df

# =========================
# 🌡️ LOAD TX
# =========================
def load_tx(file):
    start = find_data_start(file)
    df = pd.read_csv(file, skiprows=start)
    df.columns = df.columns.str.strip()

    df["DATE"] = pd.to_datetime(df["DATE"], format="%Y%m%d")
    df["year"] = df["DATE"].dt.year

    df["TX"] = pd.to_numeric(df["TX"], errors="coerce") / 10.0
    df = df[df["TX"] != -999.9]

    return df

# =========================
# 📂 LOAD FILES
# =========================
rr_files = glob.glob("RR_STAID*.txt")
tg_files = glob.glob("TG_STAID*.txt")
tn_files = glob.glob("TN_STAID*.txt")
tx_files = glob.glob("TX_STAID*.txt")

print("RR:", len(rr_files), "TG:", len(tg_files), "TN:", len(tn_files), "TX:", len(tx_files))

# =========================
# LOAD ALL DATA
# =========================
df_rr = pd.concat([load_rr(f) for f in rr_files])
df_tg = pd.concat([load_tg(f) for f in tg_files])
df_tn = pd.concat([load_tn(f) for f in tn_files])
df_tx = pd.concat([load_tx(f) for f in tx_files])

# =========================
# 🌧️ RAIN FEATURES
# =========================

# Total yearly rainfall per station → avg across stations
prcp = df_rr.groupby(["year", "SOUID"])["RR"].sum().groupby("year").mean()

# Rainy days (FIXED threshold: >1 mm)
rd = df_rr[df_rr["RR"] > 1].groupby(["year", "SOUID"]).size().groupby("year").mean()

rain_features = pd.DataFrame({
    "PRCP": prcp,
    "RD": rd
}).fillna(0)

# =========================
# 🌡️ CORE TEMP FEATURES
# =========================
tmean = df_tg.groupby(["year", "SOUID"])["TG"].mean().groupby("year").mean()
tmin = df_tn.groupby(["year", "SOUID"])["TN"].mean().groupby("year").mean()
tmax = df_tx.groupby(["year", "SOUID"])["TX"].mean().groupby("year").mean()

# =========================
# 🌡️ DERIVED FEATURES
# =========================

# Temperature variability
tvar = tmax - tmin

# Frost days
fd = df_tg[df_tg["TG"] < 0].groupby(["year", "SOUID"]).size().groupby("year").mean()

# Hot days
hd = df_tx[df_tx["TX"] > 30].groupby(["year", "SOUID"]).size().groupby("year").mean()

# =========================
# 🌱 GDD (IMPROVED VERSION)
# =========================
# Cap at 30°C (biological realism)
df_tg["TG_capped"] = df_tg["TG"].clip(upper=30)

df_tg["GDD"] = (df_tg["TG_capped"] - 10).clip(lower=0)

gdd = df_tg.groupby(["year", "SOUID"])["GDD"].sum().groupby("year").mean()

# =========================
# 🔥 STRESS FEATURES
# =========================

# Heat stress (>35°C)
heat_stress = df_tx[df_tx["TX"] > 35].groupby(["year", "SOUID"]).size().groupby("year").mean()

# Cold stress (<0°C)
cold_stress = df_tn[df_tn["TN"] < 0].groupby(["year", "SOUID"]).size().groupby("year").mean()

# =========================
# 🔗 FINAL MERGE
# =========================
final_climate = rain_features.join(pd.DataFrame({
    "Tmean": tmean,
    "Tmin": tmin,
    "Tmax": tmax,
    "Tvar": tvar,
    "FD": fd,
    "HD": hd,
    "GDD": gdd,
    "Heat_Stress": heat_stress,
    "Cold_Stress": cold_stress
}), how="inner").reset_index()

# =========================
# 🧪 SANITY CHECK
# =========================
print("\n=== FINAL SANITY CHECK ===")
print("PRCP:", final_climate["PRCP"].min(), "-", final_climate["PRCP"].max())
print("RD:", final_climate["RD"].min(), "-", final_climate["RD"].max())
print("Tmean:", final_climate["Tmean"].min(), "-", final_climate["Tmean"].max())
print("GDD:", final_climate["GDD"].min(), "-", final_climate["GDD"].max())

# =========================
# 💾 SAVE
# =========================
final_climate.to_csv("climate_final_corrected.csv", index=False)

print("\nSaved as climate_final_corrected.csv ✅")
print(final_climate.head())
```
