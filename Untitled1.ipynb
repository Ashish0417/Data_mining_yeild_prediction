```python
import pandas as pd
import glob
import os

# =========================
# ⚙️ CONFIG
# =========================
DATA_PATH = "ECA_blended"
os.chdir(DATA_PATH)

# =========================
# 🔍 FIND DATA START
# =========================
def find_data_start(file):
    with open(file, 'r') as f:
        for i, line in enumerate(f):
            if line.strip().startswith("SOUID"):
                return i
    return None

# =========================
# 🌧️ LOAD RAIN FILE
# =========================
def load_rr_file(file):
    start = find_data_start(file)
    if start is None:
        return None

    df = pd.read_csv(file, skiprows=start)
    df.columns = df.columns.str.strip()

    # Convert date
    df["DATE"] = pd.to_datetime(df["DATE"], format="%Y%m%d")
    df["year"] = df["DATE"].dt.year

    # Convert RR (0.1 mm → mm)
    df["RR"] = pd.to_numeric(df["RR"], errors="coerce") / 10.0

    # Remove missing
    df = df[df["RR"] != -999.9]

    return df

# =========================
# 🌡️ LOAD TEMP FILE
# =========================
def load_tg_file(file):
    start = find_data_start(file)
    if start is None:
        return None

    df = pd.read_csv(file, skiprows=start)
    df.columns = df.columns.str.strip()

    df["DATE"] = pd.to_datetime(df["DATE"], format="%Y%m%d")
    df["year"] = df["DATE"].dt.year

    # Convert TG (0.1°C → °C)
    df["TG"] = pd.to_numeric(df["TG"], errors="coerce") / 10.0

    df = df[df["TG"] != -999.9]

    return df

# =========================
# 📂 LOAD FILES
# =========================
rr_files = glob.glob("RR_STAID*.txt")
tg_files = glob.glob("TG_STAID*.txt")

print("RR files:", len(rr_files))
print("TG files:", len(tg_files))

# =========================
# 🌧️ PROCESS RAIN DATA
# =========================
rr_list = []

for file in rr_files:
    df = load_rr_file(file)
    if df is not None:
        rr_list.append(df)

if len(rr_list) == 0:
    raise ValueError("No rainfall data loaded")

df_rr = pd.concat(rr_list)

# =========================
# 🌧️ RAIN FEATURES (CORRECT)
# =========================

# 1. PRCP → yearly rainfall per station → avg across stations
prcp_station = df_rr.groupby(["year", "SOUID"])["RR"].sum()
prcp = prcp_station.groupby("year").mean()

# 2. RD → rainy days per station → avg across stations
rd_station = df_rr[df_rr["RR"] > 0].groupby(["year", "SOUID"]).size()
rd = rd_station.groupby("year").mean()

rain_features = pd.DataFrame({
    "PRCP": prcp,
    "RD": rd
}).fillna(0)

# =========================
# 🌡️ PROCESS TEMP DATA
# =========================
tg_list = []

for file in tg_files:
    df = load_tg_file(file)
    if df is not None:
        tg_list.append(df)

if len(tg_list) == 0:
    raise ValueError("No temperature data loaded")

df_tg = pd.concat(tg_list)

# =========================
# 🌡️ TEMP FEATURES (CORRECT)
# =========================

# 1. Tmean → yearly avg per station → avg across stations
tmean_station = df_tg.groupby(["year", "SOUID"])["TG"].mean()
tmean = tmean_station.groupby("year").mean()

# 2. FD → frost days per station → avg across stations
fd_station = df_tg[df_tg["TG"] < 0].groupby(["year", "SOUID"]).size()
fd = fd_station.groupby("year").mean()

# 3. HD → hot days per station → avg across stations
hd_station = df_tg[df_tg["TG"] > 30].groupby(["year", "SOUID"]).size()
hd = hd_station.groupby("year").mean()

temp_features = pd.DataFrame({
    "Tmean": tmean,
    "FD": fd,
    "HD": hd
}).fillna(0)

# =========================
# 🔗 MERGE FEATURES
# =========================
final_climate = rain_features.join(temp_features, how="inner")

final_climate = final_climate.reset_index()

# =========================
# 🧪 SANITY CHECK (CRITICAL)
# =========================
print("\n=== Sanity Check ===")
print("PRCP range:", final_climate["PRCP"].min(), "-", final_climate["PRCP"].max())
print("RD range:", final_climate["RD"].min(), "-", final_climate["RD"].max())
print("Tmean range:", final_climate["Tmean"].min(), "-", final_climate["Tmean"].max())
print("FD range:", final_climate["FD"].min(), "-", final_climate["FD"].max())
print("HD range:", final_climate["HD"].min(), "-", final_climate["HD"].max())

# =========================
# 💾 SAVE
# =========================
final_climate.to_csv("climate_yearly_corrected.csv", index=False)

print("\nSaved as climate_yearly_corrected.csv ✅")
print("\nSample output:")
print(final_climate.head())
```
